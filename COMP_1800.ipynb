```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

CinemaWeeklyVisitors = pd.read_csv('https://tinyurl.com/ChrisCoDV/001354133/CinemaWeeklyVisitors.csv', index_col=0)
cinema_age = pd.read_csv('https://tinyurl.com/ChrisCoDV/001354133/CinemaAge.csv', index_col=0)
CinemaCapacity = pd.read_csv('https://tinyurl.com/ChrisCoDV/001354133/CinemaCapacity.csv',index_col=0)
CinemaMarketing = pd.read_csv('https://tinyurl.com/ChrisCoDV/001354133/CinemaMarketing.csv',index_col=0)
CinemaOverheads = pd.read_csv('https://tinyurl.com/ChrisCoDV/001354133/CinemaOverheads.csv',index_col=0)
CinemaSpend = pd.read_csv('https://tinyurl.com/ChrisCoDV/001354133/CinemaSpend.csv',index_col=0)

summary_data = pd.DataFrame(index=CinemaWeeklyVisitors.columns)
summary_data['Avg age (yrs)'] = cinema_age.values
summary_data['Seating capacity'] = CinemaCapacity.values
summary_data['WeeklyVisitors'] = CinemaWeeklyVisitors.sum().values
summary_data['Marketing (£000s)'] = CinemaMarketing.values
summary_data['Overheads (£000s)'] = CinemaOverheads.values
summary_data['Avg spend (£)'] = CinemaSpend.values
print(summary_data.head())
```

```python
#DISTRIBUTION OF WEEKLY VISITORS AT EACH CINEMA
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = "https://tinyurl.com/ChrisCoDV/001354133/CinemaWeeklyVisitors.csv"
cinema_weekly_visitors = pd.read_csv(file_path)

# Plotting the distribution of weekly visitors at each cinema
plt.figure(figsize=(12, 8))
sns.barplot(data=cinema_weekly_visitors, orient='h')
plt.title('Distribution of Weekly Visitors at Each Cinema')
plt.xlabel('Number of Weekly Visitors')
plt.ylabel('Cinema')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=summary_data, x='Marketing (£000s)', y='Avg spend (£)')
plt.title('Marketing Spend vs. Average Spend')
plt.xlabel('Marketing Spend (£000s)')
plt.ylabel('Average Spend (£)')
plt.show()
```

```python
#1
plt.figure(figsize=(10, 6))
sns.barplot(data=summary_data, x=summary_data.index, y='Avg age (yrs)')
plt.title('Average Age of Visitors by Cinema')
plt.xlabel('Cinema')
plt.ylabel('Average Age (years)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(summary_data['WeeklyVisitors'], bins=20, kde=True)
plt.title('Distribution of Weekly Visitors')
plt.xlabel('Weekly Visitors')
plt.ylabel('Frequency')
plt.show()
```

```python
#3
plt.figure(figsize=(10, 6))
sns.barplot(data=summary_data, x=summary_data.index, y='Marketing (£000s)')
plt.title('Marketing Expenditure')
plt.xlabel('Cinema')
plt.ylabel('Marketing Expenditure (£000s)')
plt.xticks(rotation=45)
plt.show()
```

```python
#5
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(summary_data.index, summary_data['WeeklyVisitors'], marker='o')
plt.title('Weekly Visitors Over Time')
plt.xlabel('Cinema')
plt.ylabel('Weekly Visitors')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#2
plt.figure(figsize=(12, 6))
summary_data.iloc[:-5]['Overheads (£000s)'].plot(marker='o')
plt.title('Overheads Over Time')
plt.xlabel('Cinema')
plt.ylabel('Overheads (£000s)')
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(summary_data, x='Seating capacity', y='Avg spend (£)', title='Seating Capacity vs. Average Spend')
fig.update_layout(xaxis_title='Seating Capacity', yaxis_title='Average Spend (£)')
fig.show()
```

```python
import plotly.express as px

fig = px.bar(summary_data, x=summary_data.index, y='Seating capacity', title='Seating Capacity for Each Cinema')
fig.update_layout(xaxis_title='Cinema', yaxis_title='Seating Capacity')
fig.show()
```

```python
import plotly.express as px

fig = px.bar(summary_data, x=summary_data.index, y=['Marketing (£000s)', 'Overheads (£000s)'],
             title='Marketing Expenditure vs Overheads for Each Cinema', barmode='stack')
fig.update_layout(xaxis_title='Cinema', yaxis_title='Amount (£000s)')
fig.show()
```

```python
#7
import plotly.express as px

fig = px.box(summary_data, x=summary_data['WeeklyVisitors'].apply(lambda x: 'High' if x > 100000 else ('Medium' if 50000 <= x <= 100000 else 'Low')),
             y='Avg spend (£)', title='Average Spend for Different Volume Cinemas')
fig.update_layout(xaxis_title='Volume', yaxis_title='Average Spend (£)')
fig.show()
```

```python
#8
import plotly.express as px

fig = px.scatter(summary_data, x='WeeklyVisitors', y='Avg spend (£)', text=summary_data.index,
                 title='Weekly Visitors vs. Average Spend with Cinema Labels')
fig.update_traces(textposition='top center')
fig.update_layout(xaxis_title='Weekly Visitors', yaxis_title='Average Spend (£)')
fig.show()
```

```python
#4
import matplotlib.pyplot as plt

# Detect anomalies based on average spend
anomalies = summary_data[(summary_data['Avg spend (£)'] < summary_data['Avg spend (£)'].quantile(0.05)) |
                         (summary_data['Avg spend (£)'] > summary_data['Avg spend (£)'].quantile(0.95))]

# Plot anomalies
plt.figure(figsize=(10, 6))
plt.scatter(summary_data.index, summary_data['Avg spend (£)'], label='Average Spend (£)')
plt.scatter(anomalies.index, anomalies['Avg spend (£)'], color='red', label='Anomalies')
plt.title('Anomalies Detection in Average Spend')
plt.xlabel('Cinema')
plt.ylabel('Average Spend (£)')
plt.legend()
plt.show()
```
